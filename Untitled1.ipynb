```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, roc_curve, auc, accuracy_score
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
import statsmodels.api as sm
from sklearn.metrics import roc_auc_score, classification_report
from imblearn.over_sampling import SMOTE
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import joblib
```

```python
df= pd.read_csv("../data/healthcare-dataset-stroke-data.csv")
```

```python
df.head()
```

```python
missing_ratio = df.isna().mean()
missing_ratio
```

```python
df = df.drop(columns=["id"])
df["bmi"] = df["bmi"].fillna(df["bmi"].median())
```

```python
for col in df.columns:
    print(f"{col} has {len(df[col].unique())} unique values and is of type {df[col].dtype}")
```

```python
df["ever_married"] = df["ever_married"].map({"No": 0, "Yes": 1})
df["Residence_type"] = df["Residence_type"].map({"Rural": 0, "Urban": 1})
df = pd.get_dummies(df, columns=["gender"], drop_first=True)
df["smoking_status"] = df["smoking_status"].replace("Unknown", "missing")
df = pd.get_dummies(df, columns=["smoking_status"], drop_first=True)
df = pd.get_dummies(df, columns=["work_type"], drop_first=True)
```

```python
df = df.astype(float)
```

```python
print((df['stroke'] == 1).sum())
print((df['stroke'] == 0).sum())
```

```python
X = df.drop(columns = ['stroke'])
y = df['stroke']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)
```

```python
X_sm = sm.add_constant(X_train)

model = sm.Logit(y_train, X_sm).fit()

print(model.summary())
```

```python
model = RandomForestClassifier(
    class_weight="balanced",
    random_state=17
)
model.fit(X_train, y_train)
y_prob = model.predict_proba(X_test)[:, 1]

y_pred = (y_prob > 0.2).astype(int)

print("ROC-AUC:", roc_auc_score(y_test, y_prob))
print(classification_report(y_test, y_pred))
```

```python
model = RandomForestClassifier(
    class_weight="balanced",
    random_state=17
)
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

model.fit(X_train_res, y_train_res)
y_prob = model.predict_proba(X_test)[:, 1]

y_pred = (y_prob > 0.2).astype(int)

print("ROC-AUC:", roc_auc_score(y_test, y_prob))
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
